* **Usando o operador +**: Voc√™ pode usar o operador + para concatenar strings

###1.2.2. Indexa√ß√£o de Fatiamento de Strings

Um dos recursos poderosos das strings em Python √© a capacidade de trabalhar com fatiamento de strings. As fatias de string permitem extrair substrings de uma determinada string especificando um intervalo de √≠ndices.

Para criar uma fatia de string, utiliza-se a seguinte sintaxe:

`string[inicio:fim:step]`

*   inicio: O √≠ndice no qual a fatia come√ßa. √â inclusivo, o que significa que o caractere neste √≠ndice est√° inclu√≠do na fatia.
* fim: o √≠ndice no qual a fatia termina. √â exclusivo, o que significa que o caractere neste √≠ndice n√£o est√° inclu√≠do na fatia.
* step (opcional): O valor do passo indica quantos caracteres pular entre cada caractere na fatia. O valor padr√£o √© 1, o que significa que caracteres consecutivos est√£o inclu√≠dos.

Abaixo seguem alguns exemplos de fatiamemto de strings

````python
nome = 'Joao da Silva Sauro'

#Imprime do primeiro ao s√©timo caractere
print(nome[0:58])

#Imprime do quinto ao s√©timo caractere
print(nome[5:8])

#Imprime do primeiro ao s√©timo caractere, com step de 2 caracteres.
print(nome[0:8:2])
````

Pode-se, ainda, omitir o inicio e o fim. No caso da omiss√£o do valor de inicio, tem-se por padr√£o o valor 0. Em caso de omiss√£o do valor de fim, tem-se por padr√£o o final da string.

````python
nome = 'Joao da Silva Sauro'

#Imprime do primeiro ao s√©timo caractere
print(nome[:8])

#Imprime do quinto ao √∫ltimo caractere
print(nome[5:])
````

H√°, ainda, a op√ß√£o de se utilizar √≠ndices negativos. Os √≠**ndices negativos **em Python s√£o um recurso conveniente que permite acessar elementos do final e n√£o do in√≠cio. A indexa√ß√£o negativa come√ßa em -1 para o √∫ltimo elemento, -2 para o pen√∫ltimo elemento e assim por diante, retrocedendo na sequ√™ncia.

Por exemplo, se voc√™ tiver uma string "Python", poder√° acessar o √∫ltimo caractere 'n' usando -1 como √≠ndice:

````python
nome = 'Python'

#imprima do primeiro ao penultimo caractere
print(nome[:-1])

#extraia os ultimos dois caracteres
print(nome[-2:])
````

### Operadores aritm√©ticos b√°sicos

As opera√ß√µes aritm√©ticas s√£o opera√ß√µes matem√°ticas fundamentais realizadas em tipos de dados num√©ricos em Python. Essas opera√ß√µes permitem manipular e calcular valores num√©ricos. Dentre as principais opera√ß√µes, destacamos:

1. **Adi√ß√£o (+)**: O operador (+) soma dois ou mais n√∫meros. √â usado apenas para adicionar n√∫meros inteiros e flutuantes, mas tamb√©m para concatenar strings. a + b
2. **Subtra√ß√£o (-)**: O operador (-) subtrai o operando direito do operando esquerdo. a - b
3. **Multiplica√ß√£o (*)**: O operador (*) multiplica dois ou mais n√∫meros. a * b
4. **Divis√£o (/)**: O operador de divis√£o (/) realiza a divis√£o entre o operando esquerdo e o operando direito. No Python 3, ele retorna um float mesmo para divis√£o inteira. a / b

O Python segue a ordem padr√£o de opera√ß√µes, o que significa que as opera√ß√µes entre par√™nteses s√£o realizadas primeiro, seguidas de exponencia√ß√£o, multiplica√ß√£o e divis√£o (da esquerda para a direita) e, finalmente, adi√ß√£o e subtra√ß√£o (da esquerda para a direita).
