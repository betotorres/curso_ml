## 2.1 Comandos Básicos de Entrada e Saída

A estrutura básica de um programa geralmente envolve uma entrada de dados, a execução de algum tipo de processamento e a devolução de uma saída. Na linguagem de programação Python, as funções input() e print() são as mais básicas para cumprir as funções de entrada e saída, respectivamente. Essas duas funções são essenciais para interagir com os usuários, exibir informações e coletar entradas do usuário durante a execução do programa.

A função input() é usada para coletar informações do usuário por meio do console. Quando usada, é exibido um prompt para que o usuário insira o valor de entrada. No colab, após a digitação do valor de entrada, o mesmo é impresso na tela.

Para visualizar o funcionamento do método, execute a célula a seguir.

```python
input()

```

Em sua sintaxe, o comando input() permite não apenas coletar os dados de entrada, mas exibir ao usuário uma mensagem informando o que é esperado de entrada. Para isso, basta colocar a mensagem como parâmetro de entrada do método.

Parâmetros são elementos que são utilizados em uma função quando a mesma é utilizada, ou chamada. Em python, o valor dos parâmetros é informado entre os parênteses existentes após o nome da função. No caso da função input(), por exemplo, os parâmetro seria posicionado da seguinte forma:

```python
input('Este é o texto passado como parâmetro ')
```

Observe que, conforme comentado, o texto informado entre os parênteses foi apresentado como um rótulo antes da exibição da caixa de entrada dos dados ao usuário.


Já a função print() é usada para exibir valores para a saída padrão, que normalmente é o console ou terminal, a partir de um valor de entrada. Assim como no método anteriormente apresentado, esses valores de entrada podem ser fornecidos à função através do uso de parâmetros.

```python
print('Apresente esse texto na tela')
```

O resultado de uma função também pode ser utilizado como entrada de outra função. Observe o exemplo abaixo onde há o uso básico de ambas as funções - print() e input(). No comando, um valor de dados capturado através do comando input() é exibido na tela a partir do comando print().

```python
print(input("Digite o Seu nome Completo"))


```

##2.2 - Variáveis

Observe que, segundo a nossa lógica de Entrada-Processamento-Saída, algo ficou faltando no exemplo anterior, já que apenas obtivemos a entrada e a apresentamos na saída, sem que qualquer outra operação de processamento tenha sido executada.

 Para sanar esse problema, imagine que, agora, é solicitado ao usuário não mais o seu nome completo, mas o primeiro e o último nome, em requsições diferentes de entrada. Como resultado final, é esperado que seja apresentado na tela, e na mesma linha, o nome completo. Como executaria essa ação? Observe que, utilizando a estratégia anterior, não vamos obter sucesso.








```python
print(input("Digite o seu primeiro nome"))
print(input("Digite o seu segundo nome"))
```

Em linguagens de programação, quando precisamos armazenar valores que precisaremos utilizar mais a frente, indepentendemente do que iremos fazer com ele, utilizamos uma **variável**. Na prática, uma variável é uma estrutura que tem por objetivo armazenar informação que pode ser utilizada ou referenciada pelo programa que está sendo desenvolvido. Por trás dos panos, as variáveis são espécies de nomes simbólicos que utilizamos para referenciar locais da memória do computador utilizados ​​para armazenar valores em um programa de computador. A operação de "armazenar" um dado valor em uma variável é chamado da de "atribuição".

##2.2.1. Atribuição de Variáveis

As variáveis estão no centro das linguagens de programação, servindo como repositórios dinâmicos para armazenar e manipular dados. Elas desempenham um papel crucial em tornar os programas flexíveis, interativos e capazes de lidar com diversas tarefas.

As variáveis atuam como locais de armazenamento nomeados que podem conter vários tipos de dados, como números, texto ou estruturas complexas. Essa capacidade de armazenar dados permite que os programas se lembrem das informações e as utilizem posteriormente na execução Assim, ao atribuir uma variável  em Python, você está criando no rótulo a esquerda uma referência ao elemento mostrado a direita do sinal de igual.

```python
#exemplo de atribuição de um valor a uma variável em python
primeiro_nome = 'João'
idade = 13
```

No exemplo acima, criamos duas varíaveis, a primeira, chamada de primeiro_nome, armazena o valor 'João', que é um valor textual. Já a segunda, chamada de idade, armazena o valor numérico 13. Observe que, em python, os números são informados sem a necessidade de aspas, enquanto as palavras podem ser informadas utilizando-se aspas simples ou duplas.

Para exibir o valor armazenado na variável, podemos utilizar o método print, conforme já apresentrado.

```python
print(primeiro_nome)
```

Em complemento, a sintaxe do comando print() permite a apresentação do valor não apenas de uma, mas de múltiplas variáveis. Para isso, cada uma das variáveis a ser impressa deve ser adicionada como parâmetro do comando, separadas por vírgulas.

```python
print(primeiro_nome, idade)
```

As variáveis contribuem para a natureza dinâmica dos programas, permitindo que os valores mudem durante a execução. Essa adaptabilidade é essencial para a criação de software interativo que responda às entradas do usuário e evolua à medida que o programa é executado.

No exemplo abaixo, utilizados os três conceitos apresentados até agora para construção de um programa interativo - a entrada de dados com imput(), o armazenamento do valor em uma variável, e a impressão do valor de saída com print().

```python
#exemplo de atribuição multipla de variáveis
primeiro_nome = input('Digite o Seu primeiro nome: ')
print(primeiro_nome)
```

##2.2.3 Boas práticas na criação de Variáveis

Ao se criar variáveis, algumas regras devem ser observadas como, por exemplo:

* Os nomes de variáveis devem conter apenas letras, números e "underlines".
* Variáveis podem ser iniciadas apenas por uma letra ou por underline.
* Não utilize palavras chaves da linguagem como variáveis.

Além disto, algumas boas práticas devem ser seguidas, a exemplo de:

*  Os nomes das variáveis devem ser curtos, mas descritivos.
*  Os nomes das variáveis costumam ser escritos em letras minúsculas, com palavras separadas por sublinhados conforme necessário para melhorar a legibilidade.
*  Tenha cuidado ao usar a letra l minúscula e a letra O maiúscula, pois elas podem ser confundidas com os números 1 e 0.

#Seção 3 -Tipos e Estruturas de Dados

##3.1. Tipos dos Dados
As variáveis em python não precisam de declaração explícita de tipo, mas se associam ao tipo do dado informado no momento da associação do rótulo ao valor. O comando `type` permite a exibição do tipo de dado associado a uma variável.

```python
primeira_entrada = 1
type(primeira_entrada)
```

É importante ressaltar que uma determinada referência associada a uma variável em python é sobrescrita no caso de uma nova associação. Se o novo valor associado pertence a um outro tipo de dado, a variável, quando da nova atribuição, passa a exibir o tipo associado ao novo elemento, sem que, para isso, seja preciso qualquer estratégia de conversão de dados (cast).


```python
primeira_entrada = 'Joao'
type(primeira_entrada)
```

No exemplo apresentado anteriormente, foi atribuída uma nova referência a variável primeira_entrada, sem que houvesse necessidade de conversão de dados. Quando associamos o valor 1, a variável foi definida automaticamente como um número inteiro, na sequência, quando informamos o valor 'Joao', o tipo da variável foi alterado para string.
